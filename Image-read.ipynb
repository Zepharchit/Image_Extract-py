```python
#Using urllib.request for downloading an image from a url
#Using image library of Pillow for changing the dimensions
import urllib.request
import random
from PIL import Image
import os
import base64
```

```python
#fuctio to download an image usig urllib
def download_img(url):
    name = random.randrange(1,1000)#Have named the files randomly in range of 1 to 1000
    file_name = str(name) + '.jpg'
    urllib.request.urlretrieve(url,file_name)
    return file_name #returning file name for using it as the file path for sizechanging function
```

```python
#function to change the size of the image
def sizeChange(filepath):
    imge = Image.open(filepath)
    imge = imge.resize((320,528))
    imge.save(filepath)
    #saving the changes in image
    with open(filepath,"rb") as f:
        encode_image = base64.b64encode(f.read())
    #encode_image = base64.b64encode(imge)
    js_file = {'image_64':encode_image,'size':os.stat(filepath).st_size,'resolution':str(imge.size[0])+"x"+str(imge.size[1])}
    print(js_file)
```

```python
#calling the functions
url = input('Enter image URL to Download: ')
x = download_img(url)
sizeChange(x)
```
